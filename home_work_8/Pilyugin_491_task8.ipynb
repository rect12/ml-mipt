```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    
    # для Sigma нужно искать спектр, для F.T нужно cингулярное разложение
    U, d, V = np.linalg.svd(F.T)
    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot * 100.0 for i in range(n)])
    m = np.argmax(var_i > varRetained*100)
    print ('{:.2f} variance retained in {} dimensions'.format(var_i[m], m))

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = V.T[:, 0:m]
    G = np.dot(F.T, U_reduced)

    return G, U_reduced
```

```python
X, y = load_digits(return_X_y=True)
labels_true = y

from matplotlib import pyplot

pyplot.figure(1, figsize=(2, 2))
pyplot.axis('off')
pyplot.imshow(X[1].reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')
pyplot.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.manifold import TSNE

tsne_model = TSNE(n_components=2)
tsne_x = tsne_model.fit_transform(X)
```

```python
print(tsne_x.shape)
```

```python
def dbscan_with_inf(data, epsilon, min_samples):
    db = DBSCAN(eps=epsilon, min_samples=min_samples).fit(data)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

    print('Estimated number of clusters: %d' % n_clusters_)
    print("Adjusted Rand Index: %0.3f"
          % metrics.adjusted_rand_score(labels_true, labels))
    print("Adjusted Mutual Information: %0.3f"
          % metrics.adjusted_mutual_info_score(labels_true, labels))
    print("Number of noise points: %d" %db.labels_[db.labels_ == -1].size)
    return core_samples_mask, labels, n_clusters_, \
            metrics.adjusted_rand_score(labels_true, labels), \
            metrics.adjusted_mutual_info_score(labels_true, labels)
```

```python
def plot_clusters(core_samples_mask, data, labels):
    unique_labels = set(labels)
    colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
    plt.figure(1, figsize=(10, 10))
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = 'k'

        class_member_mask = (labels == k)

        xy = data[class_member_mask & core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=14)

        xy = data[class_member_mask & ~core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=6)

    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    plt.title('Estimated number of clusters: %d' % n_clusters_)
```

```python
epsilon_arr = np.linspace(1, 2, 30)
samples_arr = list(range(6, 19))
max_score = 0
super_epsilon = 0
super_samples = 0

for epsilon in epsilon_arr:
    for sample in samples_arr:
        print('Eps = {}, sample = {}'.format(epsilon, sample))
        core_samples_mask, labels, n_clusters_, adjusted_rand, adjusted_mutual = \
                                            dbscan_with_inf(tsne_x, epsilon, sample)
        if adjusted_rand + adjusted_mutual > max_score:
            super_epsilon = epsilon
            super_samples = sample
            max_score = adjusted_rand + adjusted_mutual
        print('\n')
```

```python
print('best epsilon: {}, best min_sample: {}'.format(super_epsilon, super_samples))
```

```python
core_samples_mask, labels, n_clusters_, adjusted_rand, adjusted_mutual = \
                                        dbscan_with_inf(tsne_x, super_epsilon, super_samples)
plot_clusters(core_samples_mask, tsne_x, labels)
```

```python
def show_image(image, figsize=7, shape1=275, shape2=400, shape3=3):
    pylab.figure(figsize=(figsize, figsize))
    pylab.axis('off')
    imshow(image.reshape(shape1, shape2, shape3))
    pylab.show()
```

```python
def new_image(image, k, clusters_arr):
    new_image = np.copy(image)
    for cluster_number in range(k):
        mask_class = (clusters_arr == cluster_number)
        new_color = np.mean(image[mask_class], axis=0)
        new_image[mask_class] = new_color
        
    return new_image
```

```python
image = plt.imread('two-lorikeet-birds-2293918.jpg')
shape_image = image.shape
print(image.shape)
image = image.reshape(-1, 3)
show_image(image)
```

```python
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering
```

```python
k = [2, 3, 10, 20]
k_means = []
mean_shift = []
agglomerative_clustering = []
```

```python
for num_clusters in k:
    model = KMeans(n_clusters=num_clusters, random_state=0)
    k_means.append(model.fit_predict(image))
```

```python
from sklearn.cluster import estimate_bandwidth

quantiles = [0.2, 0.16, 0.05, 0.03]

print('size of clusters:')

for quantile in quantiles:
    bandwidth = estimate_bandwidth(image, quantile=quantile, n_samples=1500)
    ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    mean_shift.append(ms.fit_predict(image))
    print(len(set(mean_shift[-1])))
```

```python
from sklearn.neighbors import kneighbors_graph

for num_clusters in k:
    connectivity = kneighbors_graph(image, 30)
    model = AgglomerativeClustering(connectivity=connectivity, n_clusters=num_clusters)
    agglomerative_clustering.append(model.fit_predict(image))
```

```python
k_size = len(k)
print('KMeans:')

for ind in range(k_size):
    print('Number of clusters: {}'.format(k[ind]))
    changed_image = new_image(image, k[ind], k_means[ind])
    print('ssim: {}'.format(compare_ssim(image.reshape(shape_image), changed_image.reshape(shape_image), multichannel=True)))
    show_image(changed_image, 5)
```

```python
q_size = len(quantiles)
print('MeanShift:')

for ind in range(q_size):
    n_clusters = len(set(mean_shift[ind]))
    changed_image = new_image(image, n_clusters, mean_shift[ind])
    print('Number of clusters: {}'.format(n_clusters))
    print('ssim: {}'.format(compare_ssim(image.reshape(shape_image), changed_image.reshape(shape_image), multichannel=True)))
    show_image(changed_image, 5)
```

```python
k_size = len(k)
print('AgglomerativeClustering:')

for ind in range(k_size):
    print('Number of clusters: {}'.format(k[ind]))
    changed_image = new_image(image, k[ind], agglomerative_clustering[ind])
    print('ssim: {}'.format(compare_ssim(image.reshape(shape_image), changed_image.reshape(shape_image), multichannel=True)))
    show_image(changed_image, 5)
```

K-means показал наилучший результат, после него идет AgglomerativeClustering.
